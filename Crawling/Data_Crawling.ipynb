#### 코노스바 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
from tqdm import tqdm

# Chrome WebDriver 설정
driver = webdriver.Chrome()

# 크롤링할 URL
blog_url = 'https://m.blog.naver.com/PostList.nhn?blogId=znlrnldnpr&categoryName=%EC%BD%94%EB%85%B8%EC%8A%A4%EB%B0%94%21&categoryNo=12&listStyle=card&logCode=0&tab=1'
driver.get(blog_url)
time.sleep(3)

# 페이지 끝까지 스크롤링하는 함수
def scroll_to_bottom(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# 스크롤 끝까지 내림
scroll_to_bottom(driver)

# 포스트 요소 찾기
posts = driver.find_elements(By.CLASS_NAME, 'card__reUkU')

# URL을 추출하여 리스트에 저장
post_urls = []
for post in posts:
    try:
        url = post.find_element(By.TAG_NAME, 'a').get_attribute('href')
        post_urls.append(url)
    except Exception as e:
        print(f"Error extracting URL from post: {e}")
        
# 종료
driver.quit()
```

```python
# 각 포스트 URL에 접속해서 제목과 내용을 추출
driver = webdriver.Chrome()  # 새로운 세션 시작
post_data = []

for url in tqdm(post_urls, desc="Crawling posts"):
    try:
        driver.get(url)
        time.sleep(2)
        title = driver.find_element(By.CLASS_NAME, 'tit_h3').text
        content = driver.find_element(By.ID, 'viewTypeSelector').text

        # 추출된 제목과 내용을 딕셔너리로 저장
        post_data.append({"title": title, "content": content})

    except Exception as e:
        print(f"Error crawling post {url}: {e}")

# 종료
driver.quit()
```

#### 폭염 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
from tqdm import tqdm

# Chrome WebDriver 설정
driver = webdriver.Chrome()

# 크롤링할 URL
blog_url = 'https://re-zero.tistory.com/444'
driver.get(blog_url)
time.sleep(2)

# 포스트 URL을 미리 추출해서 저장
post_urls = [post.get_attribute('href') for post in driver.find_elements(By.CLASS_NAME, 'tx-link')]

# 포스트 내용 추출을 위한 리스트
post_data = []

# 각 포스트 URL로 접속해서 내용 추출 후 이전 페이지로 돌아오기
for index, post_url in enumerate(post_urls):
    try:
        # 포스트 페이지로 이동
        driver.get(post_url)
        time.sleep(3)  # 페이지 로딩을 기다림

        ## 내용 크롤링
        content = driver.find_element(By.CLASS_NAME, 'article').text

        # 추출된 내용을 딕셔너리로 저장
        post_data.append({"content": content})

        # 페이지 로딩 대기 (필요 시)
        time.sleep(2)

    except Exception as e:
        print(f"Error processing post {index}: {e}")
        
# 종료
driver.quit()
```

```python
# 변환할 JSON 파일 경로
output_json_file = "이멋진폭염_텍본.json"

# 리스트를 JSON 형식으로 변환하여 파일로 저장
with open(output_json_file, 'w', encoding='utf-8') as json_file:
    json.dump(post_data, json_file, ensure_ascii=False, indent=0)
```

#### 조아라 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
from tqdm import tqdm

# Chrome WebDriver 설정
driver = webdriver.Chrome()

# 크롤링할 URL
blog_url = 'https://www.joara.com/book/1559023?tab=chapter'
driver.get(blog_url)
time.sleep(3)

# 페이지 끝까지 스크롤링하는 함수
def scroll_to_bottom(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# 스크롤 끝까지 내림
scroll_to_bottom(driver)

# 포스트 요소 찾기
posts = driver.find_elements(By.CLASS_NAME, 'episode-items')

# URL을 추출하여 리스트에 저장
post_urls = []
for post in posts:
    try:
        url = post.find_element(By.TAG_NAME, 'a').get_attribute('href')
        post_urls.append(url)
    except Exception as e:
        print(f"Error extracting URL from post: {e}")
```

```python
# 각 포스트 URL에 접속해서 제목과 내용을 추출
driver = webdriver.Chrome()  # 새로운 세션 시작
post_data = []

for url in tqdm(post_urls, desc="Crawling posts"):
    try:
        driver.get(url)
        time.sleep(2)

        WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//img[@src='/assets/ic/main_popup_close.png']"))
        ).click()
        
        content = driver.find_element(By.CLASS_NAME, 'text-wrap no-print').text

        # 추출된 제목과 내용을 딕셔너리로 저장
        post_data.append({"content": content})

    except Exception as e:
        print(f"Error crawling post {url}: {e}")

# 종료
driver.quit()
```
